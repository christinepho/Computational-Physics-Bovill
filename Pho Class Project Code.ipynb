```python
import numpy as np
from matplotlib import pyplot as plt, rcParams
import math
from scipy.integrate import odeint
from mpl_toolkits import mplot3d
%matplotlib inline
```

```python
#Predator Prey Model
def dxdt(x,y,a,b):
    return a*x-b*x*y

def dydt(x,y,g,d):
    return g*x*y-d*y
```

```python
#Parameters of the model

a=1.0  #alpha
b=0.1  #beta
g=0.1  #gamma
d=2.0  #delta


#Since we are looking at different cases of the model we have
#Different values for beta and gamma that we are plotting
bs=[0.2,0.4,0.6,0.8]
gs=[0.2,0.4,0.6,0.8]
```

```python
N=10000 #number of steps in integration
t1=0.0  #initial time
t2=30.0 #final time
t=np.linspace(t1,t2,N)
h=(t2-t1)/N  #time step

xs=np.zeros((len(bs)*len(gs),N))
ys=np.zeros((len(bs)*len(gs),N))
```

```python
####Performing Runge-Kutta RK4 Integrating
####Generating plots for different cases of beta and gamma

count=0

for i in range(len(bs)):
    b=bs[i]
    for k in range(len(gs)):
        g=gs[k]
        
        x0=2.0
        y0=2.0

        x=np.zeros(N)
        y=np.zeros(N)

        x[0]=x0
        y[0]=y0

        for i in range(1,N):
            xi=x[i-1]
            yi=y[i-1]

            k1x=h*dxdt(xi,yi,a,b)
            k1y=h*dydt(xi,yi,g,d)


            k2x=h*dxdt(xi+k1x/2,yi+k1y/2,a,b)
            k2y=h*dydt(xi+k1x/2,yi+k1y/2,g,d)

            k3x=h*dxdt(xi+k2x/2,yi+k2y/2,a,b)
            k3y=h*dydt(xi+k2x/2,yi+k2y/2,g,d)

            k4x=h*dxdt(xi+k3x,yi+k3y,a,b)
            k4y=h*dydt(xi+k3x,yi+k3y,g,d)

            x[i]=xi+1.0/6.0*(k1x + 2*k2x  + 2*k3x + k4x)   
            y[i]=yi+1.0/6.0*(k1y + 2*k2y  + 2*k3y + k4y)
        
        xs[count]=x
        ys[count]=y
        
        count=count+1
    
```

```python
#Plotting one specific case

fig = plt.figure(figsize=(20,20))
rcParams.update({'font.size': 22})
plt.figure(figsize=(5,5))
plt.plot(t,ys[3],label='fox')
plt.plot(t,xs[3],label="rabbit")
plt.text(0,27,"a4")
plt.legend()
plt.xlabel("Time")
plt.ylim(0,30)
plt.ylabel("Population * 1000")
plt.tight_layout()

plt.show()
```
