```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
N=100  #nmax
```

```python
def Simple(x):
    sum=0.0
    for n in range(N):
        sum=sum+(-1)**n*(x)**n/(math.factorial(n))
    return sum
```

```python
def Recursion(x):
    sn=1.0
    sum=sn
    for n in np.linspace(1,N-1,N-1,dtype=int):
        sn=-sn*x/n
        sum=sum+sn
    return sum
```

```python
def Inverse(x):
    sum=0.0
    for n in range(N):
        sum=sum+(x)**n/(math.factorial(n))
    return 1/sum
```

```python
#Calculating and plotting e^-x from x=0 to x=100 for each method

x=np.linspace(0,100,101)

sim=[]
rec=[]
inv=[]
nump=[]

for i in range(len(x)):
    sim=np.append(sim,Simple(x[i]))
    rec=np.append(rec,Recursion(x[i]))
    inv=np.append(inv,Inverse(x[i]))
    nump=np.append(nump,np.exp(-x[i]))

    
plt.plot(x,sim, label="Simple",  c="red")
plt.legend()
plt.xlabel("x", fontsize=20)
plt.ylabel("exp(-x)", fontsize=20)
plt.title("Simple")
plt.show()

plt.plot(x,rec, label="Recursion" )
plt.legend()
plt.xlabel("x", fontsize=20)
plt.ylabel("exp(-x)", fontsize=20)
plt.title("Recursion")
plt.show()

plt.plot(x,inv, label="Inverse",  c="green")
plt.legend()
plt.xlabel("x", fontsize=20)
plt.ylabel("exp(-x)", fontsize=20)
plt.title("Inverse")
plt.show()

plt.plot(x,nump, label="Numpy", c="orange")
plt.legend()
plt.xlabel("x", fontsize=20)
plt.ylabel("exp(-x)", fontsize=20)
plt.title("Numpy")
plt.show()

```

```python
#Plotting the difference between each method and numpy

plt.plot(x,sim-nump, label="Simple",  dashes = [5,5])
plt.plot(x,rec-nump, label="Recursion",  dashes = [1,1])
plt.plot(x,inv-nump, label="Inverse",  dashes = [4,3])
plt.legend()
plt.xlabel("x", fontsize=20)
plt.ylabel("exp(-x)", fontsize=20)
plt.title("Difference between 3 algorithms and numpy.exp()")
plt.show()
```
