```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
#this function converts angle from degrees to radians
def convert(angle):
    angle= angle*math.pi/180
    return angle
#this function calculates the distance traveled by a projectile
def distance(theta,v0,M,R):
    G=6.674e-11
    g=G*M/(R**2)
    time=abs((2*v0*math.sin(convert(theta)))/g)
    return v0*math.cos(convert(theta))*time
```

```python
Mearth=5.9722e24
Rearth=6.3781e6

Mmars=6.4171e23
Rmars=3.3895e6

Mtitan=1.3452e23
Rtitan=2.5747e6

Mceres=9.393e20
Rceres=4.73e5

M=[Mearth,Mmars,Mtitan,Mceres]
R=[Rearth,Rmars,Rtitan,Rceres]
loc=["Earth","Mars","Titan","Ceres"]
```

```python
m= float(input("Enter the mass, in kg, of the projectile: "))
v0= float(input("Enter the initial velocity, in meter/second, of the projectile: "))
theta= float(input("Enter the initial angle, in degrees, of the projectile: "))
print("Select which planet or moon the cannon is on.")

planet=int(input("Enter 0 for Earth, 1 for Mars, 2 for Titan, or 3 for Ceres: "))
while planet<0 or planet>3:
    print("That's not an integer between 0 and 3.")
    planet=int(input("Enter 0 for Earth, 1 for Mars, 2 for Titan, or 3 for Ceres: "))


print("The projectile travels a distance...")
print(distance(theta,v0,M[planet],R[planet]))
```

```python
m=np.linspace(0.1,10.0, 10)
theta=np.linspace(0,90,10)
v0=np.linspace(1,100,10)
```

```python


for j in range(len(M)):
    d=[]
    for i in range(len(m)):
        d=np.append(d,distance(45,10,M[j],R[j]))
        


    plt.plot(m,d,label=loc[j])

plt.xlabel("Mass of Projectile in kg")
plt.ylabel("Distance Traveled in meters")
plt.title("Distance of Projectile depending on Mass of Projectile")
plt.legend()
plt.show()

print("The launch angle and initial velocity are held constant at 45 degrees and 10 m/s.")
print("The graph shows that distance traveled by a projectile does not depend on its mass.")
```
